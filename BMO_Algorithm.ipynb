```python
# -*- coding: utf-8 -*-
@author:Somavarapu Aravind
import numpy as np
import matplotlib.pyplot as plt
import random
import math
from mpl_toolkits.mplot3d import *
from matplotlib import cm

f=lambda x,y,k:((3*(1-pow(x-k,2))*(np.exp(-((pow(x-k,2)+(pow((y-k)+1,2)))))))-(10*(((x-k)/5)-(pow(x-k,3))-pow(y,5))*(np.exp(-(pow(x-k,2)+pow(y-k,2)))))-(((1/3)*(np.exp(-(pow((x-k)+1,2)+pow(y-k,2)))))))
x=np.linspace(-3,3,100)
y=np.linspace(-3,3,100)

X,Y=np.meshgrid(x,y)
num_points=3
F=f(X,Y,0)
r_x=[]
r_y=[]
#---------------------------step-1:random points---------------------
for i in range(num_points):
    r_x.append(random.uniform(-3,3))
    r_y.append(random.uniform(-3,3))
    
r_X,r_Y=np.meshgrid(r_x,r_y)
r_F=f(r_X,r_Y,0)
#points together in a list
points=[]
a=0
l=0
for i in range(num_points):
    I=a
    L=l   
    for j in range(num_points):
        p=[]
        for k in range(1):
            p.append(r_X[I][L])
            p.append(r_Y[I][L])
            p.append(r_F[I][L])
        points.append(p)
        L=L+1
    a=a+1
    l=0
# print("---------------------------------points------------------------------\t")
print("------------------------------------initial points---------------------------------------")
print("\t")
print(points)
fig=plt.figure(1)
ax = fig.gca(projection='3d') 

ax.view_init(elev=25,azim=50)
ax.plot_surface(X,Y,F,cmap=cm.spring);

ax.scatter(r_X,r_Y,r_F)  
plt.xlim(max(x),min(x))
plt.ylim(max(y),min(y))
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_zlabel('z label')
plt.show()

#-----------------------STEP-2:UV DISTRIBUTION-----------------------
distance_arr=[]
for i in range(len(points)):
    distance_int_arr=[]
    for j in range(len(points)):
        distance=math.dist(points[i],points[j])
        distance_int_arr.append(distance)
    distance_arr.append(distance_int_arr)
uv=[]
for i in range(len(distance_arr)):
    uv_dist=[]
    for j in range(len(distance_arr)):
        calc_dist=distance_arr[i][j]/(sum(distance_arr[i]))
        uv_value=(points[i][2])*calc_dist
        uv_dist.append(uv_value)
    uv.append(uv_dist)
# print("--------------distance from ith point to jth point where(j=0,1,2,....)-------------\n")
# print(uv)
# print("\t\t")

#----------------------STEP-2:L-MATE SELECTION-------------------
uv_mate=[]
for i in range(len(uv)):
    uv_order=[]
    for j in range(len(uv)):
        uv_order.append(uv[j][i])
    uv_mate.append(uv_order)
# print("-----------------uv recieved by b-fly from other butterflies.....------------------\n")
# print(uv_mate)
# print("\n\n")
uv_pos=[]
for i in range(len(uv_mate)):
    pos=uv_mate[i].index(max(uv_mate[i]))
    uv_pos.append(pos)
# print("-----------------highest recieved uv by a butterfly from......-----------------")
# print(uv_pos)
print("\t")
for i in range(len(uv_pos)):
    if(points[i][2]<uv_mate[i][(uv_pos[i])]):
        print(uv_pos[i],"-->yes..you are my l-mate")
        print("\t")
    else:
        print(uv_pos[i],"-->nooooooooo........yor are not my l-mate")
        print("\t")
        
#-----------------------STEP-3:MOVEMENT PHASE------------------------
fig2=plt.figure(2)
ax2 = fig2.gca(projection='3d') 
ax2.view_init(elev=30,azim=100)
ax2.plot_surface(X,Y,F,cmap=cm.spring);
plt.xlim(max(x),min(x))
plt.ylim(max(y),min(y))
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_zlabel('z label')
abs_points=[]
final_diff=[]
abs_final_diff=[]
for i in range(len(points)):
    diff=[]
    abs_int_points=[]
    for j in range(3):
        if(points[i][2]<uv_mate[i][(uv_pos[i])]):
            abs_diff=points[uv_pos[i]][j]-points[i][j]
            abs_final_diff=abs(abs_diff)
        else:
            abs_diff=0
        abs_int_points.append(abs_final_diff)
        diff.append(abs_diff)
    abs_points.append(abs_int_points)
    final_diff.append(diff)
    
# print(abs_points)  
# print("\t")
# print(final_diff)
print("\t")
for i in range(len(points)):
    for j in range(3):
        if(points[i][2]<uv_mate[i][(uv_pos[i])]):
            if(final_diff[i][j]==0):
                step=0
            else:
                step=(final_diff[i][j]/abs_points[i][j])
            
            points[i][j]=points[i][j]+(step*0.02)

for i in range(len(points)):
    ax2.scatter(points[i][0],points[i][1],points[i][2])
plt.show()
print("\t")
print("-------------------------------------------updated points------------------------------------")
print(points)

```
